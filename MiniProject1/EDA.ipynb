```python
# coding=utf-8
"""
data EDA
"""

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from src.config import folder
import numpy as np

print("import done!")

path = folder.get_data_path('challenger_0521.csv')
df = pd.read_csv(path, index_col=0)
```

```python
positions = ['top', 'jug', 'mid', 'adc', 'sup']
for position in positions:
    df_temp = df[df['position'] == position]
    df_temp = df_temp.drop(['position'], axis=1)
    a = sns.heatmap(df_temp.corr(), annot=True)
```

```python
a = sns.boxplot(x='position', y='kill', hue='win', data=df)
a.set_title('kill', size=15)
```

```python
a = sns.boxplot(x='position', y='death', hue='win', data=df)
a.set_title('death', size=15)
```

```python
position = 'top'
df_temp = df[df['position'] == position]
df_temp = df_temp.drop(['position'], axis=1)

f, ax = plt.subplots(figsize = (11,11))
my_df_filter = df_temp
corr = my_df_filter.corr(method = 'pearson')
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
g = sns.heatmap(corr, cmap='RdYlGn_r', vmax = 1, mask = mask, center = 0, annot = True, fmt='.2f', 
                square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title(f"[{position}] Correlation Heatmap")
```
