```python
"""
Editor : Yusang Jeon
Date : 2021-05-20

크롤링된 챌린저 라인별 전적 데이터 분석하기
"""

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from setting import folder

# 링크 참고
# https://mindscale.kr/course/python-visualization-basic/grid/
filename = 'challenger_log_rev.csv'
path_df = get_data_path(filename)
df = pd.read_csv(path_df)
df = df[df.columns[1:]] # 잘못입력된 column 제거

df_win = df[df['win'] == 1]
df_win = df_win.drop(['win'], axis=1)

df_lose = df[df['win'] == 0]
df_lose = df_lose.drop(['win'], axis=1)
```

```python
# correlation
sns.heatmap(df_win.corr(), annot=True)
```

```python
sns.heatmap(df_lose.corr(), annot=True)
```

```python
sns.boxplot(x='position', y='kill', hue='win', data=df)
```

```python
grid = sns.FacetGrid(data=df, col='position')
grid.map(sns.boxplot, "win", "kill")
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2)
ax0 = sns.boxplot("win", "kill", data=df[df['position']=='mid'], ax=ax[0,0])
ax1 = sns.boxplot("win", "death", data=df[df['position']=='mid'], ax=ax[0,1])
ax2 = sns.boxplot("win", "assist", data=df[df['position']=='mid'], ax=ax[1,0])
ax3 = sns.boxplot("win", "cs", data=df[df['position']=='mid'], ax=ax[1,1])
ax0.set_title('kill', size=15)
ax1.set_title('death', size=15)
ax2.set_title('assist', size=15)
ax3.set_title('cs', size=15)
```

```python
fig, ax = plt.subplots(nrows=5,
                      figsize=(10,10),
                      constrained_layout=True)
fig.tight_layout()

plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=1, 
                    top=1, 
                    wspace=1, 
                    hspace=0.7)


kill = sns.boxplot(x='position', y='kill', hue='win', data=df, ax=ax[0])
death = sns.boxplot(x='position', y='death', hue='win', data=df, ax=ax[1])
assist = sns.boxplot(x='position', y='assist', hue='win', data=df, ax=ax[2])
cs = sns.boxplot(x='position', y='cs', hue='win', data=df, ax=ax[3])
vision = sns.boxplot(x='position', y='vision score', hue='win', data=df, ax=ax[4])

kill.set_title('kill', size=15)
death.set_title('death', size=15)
assist.set_title('assist', size=15)
cs.set_title('cs', size=15)
vision.set_title('vision', size=15)


```
